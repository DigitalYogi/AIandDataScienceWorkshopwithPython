```python
import pandas as pd
iris = pd.read_csv('https://raw.githubusercontent.com/dia1128/AIandDataScienceWorkshopwithPython/main/Week2/Assignment/iris.csv')
print(iris.head())
```

```python
import matplotlib.pyplot as plt 
```

```python
iris.plot(kind='scatter', x='petal.length', y='petal.width') 
plt.show()
```

```python
a= iris['variety'].value_counts()
species = a.index
count = a.values
plt.bar(species,count,color = 'lightgreen')
plt.xlabel('species')
plt.ylabel('count')
plt.show()
```

```python
length_width = iris[['petal.length','petal.width','sepal.length','sepal.width']] #excluding species column
length_width.boxplot()
plt.xlabel('Flower measurements')
plt.ylabel('values')
plt.title("Iris dataset analysis")
plt.show()
```

```python
x = iris["sepal.length"]
  
plt.hist(x, bins = 20, color = "green")
plt.xlabel("Sepal_Length")
plt.ylabel("Count")
```

```python
a= iris['variety'].value_counts()
species = a.index
count = a.values
colors= ['lightblue','lightgreen','gold']

plt.pie(count, labels=species,shadow=True,
        colors=colors,autopct='%1.1f%%')
plt.xlabel('species')
plt.axis('equal')
plt.show()
```
